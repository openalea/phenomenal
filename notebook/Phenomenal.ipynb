## Binarization

```python
import alinea.phenomenal.configuration
import alinea.phenomenal.binarization
```

```python
factor_binarization = alinea.phenomenal.configuration.binarization_factor('factor_image_basic.cfg')

images_binarize_mean_shift = alinea.phenomenal.binarization.binarization(images['0962']['2013-06-22'], 
                                                                         factor_binarization, 
                                                                         methods='mean_shift')
```

### Visualisation

```python
im = cv2.cvtColor(images_binarize_mean_shift[120], cv2.COLOR_GRAY2RGB)
%matplotlib inline
from matplotlib.pyplot import imshow
import numpy 

imshow(numpy.asarray(im))


alinea.phenomenal.result_viewer.show_images([images['0962']['2013-06-22'][120], im])
```

## Repair

```python
import alinea.phenomenal.repair_processing
```

```python
repair_images = alinea.phenomenal.repair_processing.repair_processing(images_binarize_mean_shift)
```

### Visualisation

```python
img_1 = cv2.cvtColor(repair_images[120], cv2.COLOR_GRAY2RGB)
img_2 = cv2.cvtColor(images_binarize_mean_shift[120], cv2.COLOR_GRAY2RGB)  
alinea.phenomenal.result_viewer.show_images([img_2, img_1])
```

## Multi-view reconstruction

```python
import alinea.phenomenal.calibration_model
import alinea.phenomenal.multi_view_reconstruction
```

```python
calibration = alinea.phenomenal.calibration_model.Calibration.read_calibration(
    'tests/test_calibration_model')
    
images_selected = dict()
for angle in repair_images:
    if 0 <= angle <= 360:
        images_selected[angle] = repair_images[angle]

points_3d = alinea.phenomenal.multi_view_reconstruction.reconstruction_3d(
    images_selected, calibration, precision=4)
```

### Visualisation

```python
alinea.phenomenal.result_viewer.show_points_3d(points_3d, scale_factor=2)
```
