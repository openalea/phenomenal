# Segmentation 3D

## 1. Prerequisites

### 1.1 Load cloud point of voxel centers

```python
%matplotlib notebook

from alinea.phenomenal.data_access.plant_1 import plant_1_voxel_centers
from alinea.phenomenal.display import (
    show_points_3d, show_image)

voxel_size = 4
voxel_centers = plant_1_voxel_centers(voxel_size=voxel_size)

# Display it
im = show_points_3d(voxel_centers, scale_factor=voxel_size)
# Documentation image incrustation
show_image(im)
```

## 2. Segmentation

### 2.1 Segment Stem

```python
from alinea.phenomenal.segmentation_3d.maize import (
    maize_stem_segmentation)

stem_voxel, not_stem_voxel = maize_stem_segmentation(
    voxel_centers, voxel_size)
```

### 2.2 Display Result

```python
%matplotlib notebook

from alinea.phenomenal.display.multi_view_reconstruction import (
    show_list_points_3d)

im = show_list_points_3d([stem_voxel, not_stem_voxel], scale_factor=voxel_size)
# Documentation image incrustation
show_image(im)
```
